```python
data = pd.read_csv('crime_data.csv')
data.head()
```

##### Select features and target / dependent and independent variable

```python
features = [
    "Offender_Race","Offender_Gender","Offender_Age","Victim_Race","Victim_Gender","Victim_Age"
]
target = "Category"

data_subset = data[features + [target]]
data_subset
```
